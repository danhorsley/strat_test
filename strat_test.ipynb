```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("TSLA.csv", index_col='Price')
df = df.drop(["Ticker","Date"])
df['Close'] = pd.to_numeric(df['Close'],downcast='float')
```

```python
def add_signals(df : pd.DataFrame, short_win : int = 50, long_win : int = 200) -> pd.DataFrame :
    df = df.copy()
    df['short_ma'] = df['Close'].rolling(short_win).mean()
    df['long_ma'] = df['Close'].rolling(long_win).mean()
    df['signal'] = 0
    df.loc[df['short_ma'] > df['long_ma'], 'signal'] = 1
    df.loc[df['short_ma'] < df['long_ma'], 'signal'] = -1
    df['pos'] = df['signal'].shift(1).fillna(0)
    return df
```

```python
ndf = add_signals(df,50,200)
ndf = ndf.dropna()
ndf['strat_rtn'] = ndf['Close'].pct_change()
ndf['cumulative_rtn'] = (1 + ndf['strat_rtn']).cumprod()

ndf.tail(20)
```
